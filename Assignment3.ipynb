1.Create a list named ‘myList’ that has the following elements: 10, 20, 30,
‘apple’, True, 8.10:




1.   Now in the ‘myList’, append these values: 30, 40

2.After that, reverse the elements of the ‘myList’ and store that in ‘reversedList’
   



```python
mylist = [10,20,30,'apple',True,8.10]

mylist.append(30)
mylist.append(40)

reversedlist = mylist[::-1]

print("Updated mylist:",mylist)
print("Reversed list:",reversedlist)
```

4.Create another tuple named numeric_tuple consisting of only integer values 10, 20, 30, 40, 50:


1. Find the minimum value from the numeric_tuple
2. Concatenate my_tuple with numeric_tuple and store the result in r1
3.   Duplicate the tuple named my_tuple 2 times and store that in ‘newdupli’




```python
numeric_tuple = (10,20,30,40,50)

min_value = min(numeric_tuple)
print("Minimum value:",min_value)

my_tuple = (1,2,3,'apple','mango')

r1 = my_tuple + numeric_tuple
print("Concatenated tuple:",r1)

newdupli = my_tuple * 2
print("Duplicated tuple:",newdupli)

```

5.Create 2 sets with the names set1 and set2, where set1 contains
 {1,2,3,4,5} and set2 contains {2,3,7,6,1}.


1.    Perform the below operation:
2. set1 union set2
3. set1 intersection set2
4. set1 difference set2
   



```python
set1 = {1,2,3,4,5}
set2 = {2,3,7,6,1}

union_set = set1.union(set2)
print("Union of set1 and set2:",union_set)

intersection_set = set1.intersection(set2)
print("Intersection of set1 and set2:",intersection_set)

difference_set = set1.difference(set2)
print("Difference of set1 and set2:",difference_set)
```

1.Concatenation in Pandas

```python
import pandas as pd

df1 = pd.DataFrame({
    'A':['A0','A1','A2'],
    'B':['B0','B1','B2']
})

print(df1)

df2 = pd.DataFrame({
    'A':['A3','A4','A5'],
    'B':['B3','B4','B5']
})

print(df2)

result = pd.concat([df1,df2])
print(result)
```

```python
result = pd.concat([df1,df2],axis=1)
print(result)
```

```python
result =pd.concat([df1,df2], ignore_index=True)
print(result)
```

Inner Merge

```python
import pandas as pd
df1=pd.DataFrame({
    'key':['A','B','C','D'],
    'Value':[1,2,3,4]

})
print(df1)
df2=pd.DataFrame({
    'key':['B','D','E','F'],
    'Value':[5,6,7,8]

})
print(df2)

result=pd.merge(df1,df2,on='key',how='inner')
print(result)
```

```python
df1=pd.DataFrame({
    'key':['A','B','C','D'],
    'Value':[1,2,3,4]

})
print(df1)
df2=pd.DataFrame({
    'key':['B','D','E','F'],
    'Value':[5,6,7,8]

})
print(df2)

result=pd.merge(df1,df2,on='key',how='left')
print(result)
```

```python
df1=pd.DataFrame({
    'key':['A','B','C','D'],
    'Value':[1,2,3,4]

})
print(df1)
df2=pd.DataFrame({
    'key':['B','D','E','F'],
    'Value':[5,6,7,8]

})
print(df2)

result=pd.merge(df1,df2,on='key',how='right')
print(result)
```

```python
df1=pd.DataFrame({
    'key':['A','B','C','D'],
    'Value':[1,2,3,4]

})
print(df1)
df2=pd.DataFrame({
    'key':['B','D','E','F'],
    'Value':[5,6,7,8]

})
print(df2)

result=pd.merge(df1,df2,on='key',how='outer')
print(result)
```

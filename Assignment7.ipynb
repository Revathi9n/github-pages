methods, i.e., first(), second(), and final().

```python
class Sub1:
  def first(self):
    print("This is the first function from Sub1 class.")

class Sub2:
  def second(self):
    print("This is the second function from the Sub2 class.")

class Super:
  def final(self):
    print("This is the final method from the Super class.")

  def call_all(self):
    sub1_obj = Sub1()
    sub2_obj = Sub2()

    sub1_obj.first()
    sub2_obj.second()
    self.final()

super_obj = Super()
super_obj.call_all()
```

 5. Create a class named ‘Parent’, and inside the class, create a function

 * Now create a class named ‘Child’ and inside the class, create a method

 * Now create an object of Hybrid class and with the help of the object, call the ‘fun1’, ‘fun2’ and ‘fun3’ methods.

```python
class Parent:
  def fun1(self):
    print("This is the message from fun1.")

class Child:
  def fun2(self):
    print("This is the message from fun2.")

class Hybrid(Parent, Child):
  def fun3(self):
    print("This is the message from fun3.")

obj = Hybrid()

obj.fun1()
obj.fun2()
obj.fun3()
```
